#### TASK: Vẽ một hình chữ nhật trống MÀU ĐỎ quanh khuôn mặt của chú chó và hiển thị trong notebook.

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Chuyển ảnh về màu RGB tức là màu gốc
plt.imshow(RGB_img)
```

```python
# Đọc hình ảnh chú chó
img = RGB_img

# Tọa độ của khuôn mặt (x, y, width, height)
face_coordinates = [(30, 25, 130, 155)]

# Vẽ hình chữ nhật màu đỏ quanh khuôn mặt 
for (x, y, w, h) in face_coordinates:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)  

# Hiển thị hình ảnh với hình chữ nhật màu đỏ quanh khuôn mặt
plt.imshow(img)

```

#### TASK: Vẽ một HÌNH TAM GIÁC MÀU XANH DƯƠNG ở giữa ảnh. Có thể chọn kích thước và góc tùy ý, nhưng phải là hình tam giác (3 cạnh) theo bất kỳ hướng nào.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Đọc hình ảnh chó
img = cv2.imread("dog.jpg")

# Tạo hình tam giác với viền đỏ quanh khuôn mặt (nếu có)
for (x, y, w, h) in face_coordinates:
    # Tọa độ các đỉnh của tam giác
    triangle_points = np.array([(x + w // 2, y), (x, y + h), (x + w, y + h)], np.int32)
    triangle_points = triangle_points.reshape((-1, 1, 2))

    # Vẽ hình tam giác với viền đỏ
    cv2.polylines(img, [triangle_points], isClosed=True, color=(255, 0, 0), thickness=2) 

# Hiển thị hình ảnh với hình tam giác màu đỏ quanh khuôn mặt
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

### BONUS TASK. Bạn có thể tô màu tam giác này không? Điều này cần một hàm khác, không có trong bài giảng. Hãy thử xem liệu bạn có thể tìm thấy trên Google không.

[CLICK VÀO LIÊN KẾT SAU ĐỂ XEM GỢI Ý](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Đọc hình ảnh chó
img = cv2.imread("dog.jpg")

# Tạo hình tam giác với viền màu xanh cơ quanh khuôn mặt (nếu có)
for (x, y, w, h) in face_coordinates: # face_coordinates chính là hình vuông bên trên mình đã vẽ
    # Tạo tọa độ của các đỉnh của tam giác
    triangle_points = np.array([(x + w // 2, y), (x, y + h), (x + w, y + h)], np.int32)
    triangle_points = triangle_points.reshape((-1, 1, 2))

    # Vẽ hình tam giác với viền màu xanh cơ
    cv2.fillPoly(img, [triangle_points], color=(255, 0, 0))  

# Hiển thị hình ảnh với hình tam giác màu xanh cơ quanh khuôn mặt
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
